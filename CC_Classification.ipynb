```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

cc_df=pd.read_csv('Data/creditcard.csv')
cc_df
```

```python
cc_df.info()
```

```python
from sklearn.model_selection import train_test_split
train_split=round(len(cc_df)*0.8)
train_set,val_set=cc_df.iloc[:train_split,:],cc_df.iloc[train_split:,:]  #Temporal Split instead of Random Split
train_set.shape,val_set.shape
```

```python
val_set
```

```python
input_cols=['Time', 'V1', 'V2', 'V3', 'V4', 'V5', 'V6', 'V7', 'V8', 'V9', 'V10',
       'V11', 'V12', 'V13', 'V14', 'V15', 'V16', 'V17', 'V18', 'V19', 'V20',
       'V21', 'V22', 'V23', 'V24', 'V25', 'V26', 'V27', 'V28', 'Amount']
target_col='Class'
```

```python
train_inputs=train_set[input_cols].copy()
val_inputs=val_set[input_cols].copy()
train_target=train_set[target_col].copy()
val_target=val_set[target_col].copy()
```

```python
numeric_cols=train_inputs.select_dtypes(include=np.number).columns.tolist()
numeric_cols
```

```python
from sklearn.preprocessing import StandardScaler #Normalised features to have a consistency
scaler=StandardScaler()
scaler.fit(train_inputs[numeric_cols])
```

```python
train_inputs[numeric_cols]=scaler.transform(train_inputs[numeric_cols])
val_inputs[numeric_cols]=scaler.transform(val_inputs[numeric_cols])
```

```python
import mlflow
import mlflow.sklearn
```

```python
from sklearn.metrics import accuracy_score,precision_score,recall_score,roc_auc_score #precision and recall matters more than accuracy due to severe class imabalance
def model_evaluate(model):
    train_preds=model.predict(train_inputs)
    val_preds=model.predict(val_inputs)
    train_acc=accuracy_score(train_target,train_preds)
    val_acc=accuracy_score(val_target,val_preds)
    train_prec=precision_score(train_target,train_preds)   
    val_prec=precision_score(val_target,val_preds)
    train_rec=recall_score(train_target,train_preds)   
    val_rec=recall_score(val_target,val_preds)   
    train_roc=roc_auc_score(train_target,train_preds)
    val_roc=roc_auc_score(val_target,val_preds)
    print(f'Train Results: Acc ==> {train_acc}, Precision ==>{train_prec},Recall ==> {train_rec},ROC-AUC ==> {train_roc}')
    print(f'Val Results: Acc ==> {val_acc}, Precsion ==>{val_prec}, Recall ==> {val_rec},ROC-AUC ==>{val_roc}')
    mlflow.log_metric('Train_Accuracy',train_acc)
    mlflow.log_metric('Test_Accuracy',val_acc)
    mlflow.log_metric('Train_Precision',train_prec)
    mlflow.log_metric('Test_Precision',val_prec)
    mlflow.log_metric('Train_Recall',train_rec)
    mlflow.log_metric('Test_Recall',val_rec)
    mlflow.log_metric('Train_ROC-AUC',train_roc)
    mlflow.log_metric('Test_ROC-AUC',val_roc)
```

```python
mlflow.set_experiment('Credit Card Fraud Detection ==> Classification')
```

```python
from sklearn.linear_model import LogisticRegression # Using this as Baseline Model
with mlflow.start_run(run_name='Logistic_Regression_V2'):
     mlflow.set_tag('description','Earlier Random Splitting was used, now this new run has temporal split (80-20)')
     model_log=LogisticRegression(class_weight='balanced')
     model_log.fit(train_inputs,train_target)
     mlflow.log_params(model_log.get_params())
     model_evaluate(model_log)
```

```python
from sklearn.tree import DecisionTreeClassifier
with mlflow.start_run(run_name='Descision_Tree v3'):
   mlflow.set_tag('description','Earlier Random Splitting was used, now this new run has temporal split (80-20)')
   model_tree=DecisionTreeClassifier(max_depth=8,random_state=42,class_weight='balanced')
   model_tree.fit(train_inputs,train_target)
   mlflow.log_params(model_tree.get_params())
   model_evaluate(model_tree)
```

```python
from sklearn.ensemble import RandomForestClassifier
with mlflow.start_run(run_name='Random_Forest v5'):
   mlflow.set_tag('description','Earlier Random Splitting was used, now this new run has temporal split (80-20)')
   model_forest=RandomForestClassifier(n_jobs=-1,n_estimators=120,max_depth=8,random_state=42,class_weight='balanced')
   model_forest.fit(train_inputs,train_target)
   mlflow.log_params(model_forest.get_params())
   model_evaluate(model_forest)
```

```python
from xgboost import XGBClassifier
with mlflow.start_run(run_name='XGB v5'):
   from sklearn.utils.class_weight import compute_sample_weight
   sample_weight=compute_sample_weight(class_weight='balanced',y=train_target)
   mlflow.set_tag('description','Earlier Random Splitting was used, now this new run has temporal split (80-20)')
   model_xgb=XGBClassifier(n_jobs=-1,n_estimators=55,max_depth=6,sample_weight=sample_weight,random_state=42)
   model_xgb.fit(train_inputs,train_target)
   mlflow.log_params(model_xgb.get_params())
   model_evaluate(model_xgb)
```
